```python
import pandas as pd
import numpy as np
```

```python
hof_df = pd.read_csv("inductees.csv")
hof_df.drop(hof_df.loc[hof_df["Years Waited"]=="*"].index, inplace=True)
hof_df.drop(columns=["Inducted By"], inplace=True)
```

```python
hof2_df = pd.read_csv("snubs.csv")
hof2_df.head()
```

```python
hof3_df = pd.read_csv("nominees_2019.csv")
hof3_df.head()
```

```python
infl_df = pd.read_csv("The 100 most influential artists, according to AllMusic data.csv")
infl_df = infl_df.rename(columns={"artistname" : "Artist", "artists influenced" : "Influenced"})
```

```python
rnr_df = pd.merge(hof_df, infl_df, on="Artist", how="left")
rnr_df = rnr_df.sort_values(by=["Influenced"])
rnr2_df = pd.merge(hof2_df, infl_df, on="Artist", how="left")
rnr_df = rnr_df.sort_values(by=["Influenced"])
rnr3_df = pd.merge(hof3_df, infl_df, on="Artist", how="left")
rnr_df = rnr_df.sort_values(by=["Influenced"])
```

```python
uss_df = pd.read_csv("Album sales US.csv")
rnr_df = pd.merge(rnr_df, uss_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, uss_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, uss_df, on="Artist", how="left")
```

```python
rnr_df = rnr_df.sort_values(by=["Certified Units (Millions)"], ascending = False)
rnr2_df = rnr2_df.sort_values(by=["Certified Units (Millions)"], ascending = False)
rnr3_df = rnr3_df.sort_values(by=["Certified Units (Millions)"], ascending = False)
rsa_df = pd.read_csv("RS500.csv")
rnr_df = pd.merge(rnr_df, rsa_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, rsa_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, rsa_df, on="Artist", how="left")
```

```python
hot_df = pd.read_csv("hot100singles.csv")
hot_df = hot_df.rename(columns={"Row Labels" : "Artist"})
rnr_df = pd.merge(rnr_df, hot_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, hot_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, hot_df, on="Artist", how="left")
```

```python
rsc_df = pd.read_csv("RSCovers.csv")
rnr_df = pd.merge(rnr_df, rsc_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, rsc_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, rsc_df, on="Artist", how="left")
```

```python
abs_df = pd.read_csv("Album_scores.csv")
rnr_df = pd.merge(rnr_df, abs_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, abs_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, abs_df, on="Artist", how="left")
```

```python
rnr_df.fillna(0, inplace=True)
rnr2_df.fillna(0, inplace=True)
rnr3_df.fillna(0, inplace=True)
```

```python
rnr_df["In_Out"] = 1
rnr2_df["In_Out"] = 0
rnr3_df["In_Out"] = 0
```
