```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
induct = pd.read_csv('output/2019.csv')
induct.drop(['Unnamed: 0',  'Years Waited', 'Year Inducted','In_Out','Artist','Highest Position','# of Years Nominated' ], axis=1, inplace=True)
induct
```

```python
future = pd.read_csv('output/2044.csv')
future.drop(['Unnamed: 0',  'Years Waited', 'Year Inducted','In_Out','Highest Position','# of Years Nominated','Artist' ], axis=1, inplace=True)
future
```

```python
hall = pd.read_csv('output/artist_data.csv')
hall.head()
hall.drop(['Unnamed: 0', 'Years Waited', 'Year Inducted', 'Artist',  '# of Years Nominated', 'Highest Position'], axis=1, inplace=True)

hall.head()
```

```python
sns.pairplot(hall,hue='In_Out',palette='Set1')
```

```python
from sklearn.model_selection import train_test_split

X = hall.drop('In_Out', axis =1)
y = hall['In_Out']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_test.columns
```

```python
#### these result are only to 1 DECIMAL place
from sklearn.ensemble import RandomForestClassifier
classify = RandomForestClassifier()
classify.fit(X_train,y_train)
induct_prob = classify.predict_proba(induct)
ip = pd.DataFrame(induct_prob)
ip
```

```python
from sklearn.ensemble import RandomForestClassifier
classify = RandomForestClassifier(n_estimators=500)
classify.fit(X_train,y_train)
#(classify.predict_proba(X_test))
fclass = classify.predict_proba(future)
fclass1 = pd.DataFrame(fclass)
fclass1
```

```python
future2 = pd.read_csv('output/2044.csv')
future1 = future2['Artist']
#induct1 = induct1.iloc[::-1]
df = pd.DataFrame(data=future1)
df.reset_index(inplace=True)
classof2044 = pd.concat([df,fclass1], axis=1)
classof2044.columns
classof2044.drop(['index',0], axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
classify = RandomForestClassifier(n_estimators=500)
classify.fit(X_train,y_train)
#(classify.predict_proba(X_test))
decimals = classify.predict_proba(induct)
decimals = pd.DataFrame(decimals)
decimals
```

```python
induct2 = pd.read_csv('output/2019.csv')
induct1 = induct2['Artist']
#induct1 = induct1.iloc[::-1]
df = pd.DataFrame(data=induct1)
df.reset_index(inplace=True)
classof2019 = pd.concat([df,decimals], axis=1)
classof2019.columns
classof2019.drop(['index',0], axis=1)
```

```python
classof2019.to_csv("output/predictions.csv")
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
predictions = dtree.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
from sklearn import metrics

print(classification_report(y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
classify = RandomForestClassifier(n_estimators=500)
classify.fit(X_train,y_train)
#(classify.predict_proba(X_test))
print(classify.predict_proba(induct))
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=600)
rfc.fit(X_train,y_train)
predictions = rfc.predict(X_test)
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
list(zip(hall.columns, rfc.feature_importances_))
```
