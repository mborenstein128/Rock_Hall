```python
#dependencies
import pandas as pd
import numpy as np
```

```python
#import main csv
hof_df = pd.read_csv("data/inductees.csv")
hof_df.drop(hof_df.loc[hof_df["Years Waited"]=="*"].index, inplace=True)
hof_df.drop(columns=["Inducted By"], inplace=True)
```

```python
#import csv
hof2_df = pd.read_csv("data/snubs.csv")
hof2_df.head()
```

```python
#import 2019 csv
hof3_df = pd.read_csv("data/nominees_2019.csv")
hof3_df.head()
```

```python
#import 2044 csv
hof4_df = pd.read_csv("data/top100spotify.csv")
hof4_df.head()
```

```python
#import influential artists csv
infl_df = pd.read_csv("data/The 100 most influential artists, according to AllMusic data.csv")
infl_df = infl_df.rename(columns={"artistname" : "Artist", "artists influenced" : "Influenced"})
```

```python
# merge df's on artist
rnr_df = pd.merge(hof_df, infl_df, on="Artist", how="left")
rnr_df = rnr_df.sort_values(by=["Influenced"])
rnr2_df = pd.merge(hof2_df, infl_df, on="Artist", how="left")
rnr2_df = rnr2_df.sort_values(by=["Influenced"])
rnr3_df = pd.merge(hof3_df, infl_df, on="Artist", how="left")
rnr3_df = rnr3_df.sort_values(by=["Influenced"])
rnr4_df = pd.merge(hof4_df, infl_df, on="Artist", how="left")
rnr4_df = rnr4_df.sort_values(by=["Influenced"])
```

```python
# merge album sales to df's on artist
uss_df = pd.read_csv("data/Album sales US.csv")
rnr_df = pd.merge(rnr_df, uss_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, uss_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, uss_df, on="Artist", how="left")
rnr4_df = pd.merge(rnr4_df, uss_df, on="Artist", how="left")
```

```python
# sort by sales, merge Rolling Stone top 500 album data on artist
rnr_df = rnr_df.sort_values(by=["Certified Units (Millions)"], ascending = False)
rnr2_df = rnr2_df.sort_values(by=["Certified Units (Millions)"], ascending = False)
rnr3_df = rnr3_df.sort_values(by=["Certified Units (Millions)"], ascending = False)
rnr4_df = rnr4_df.sort_values(by=["Certified Units (Millions)"], ascending = False)
rsa_df = pd.read_csv("data/RS500.csv")
rnr_df = pd.merge(rnr_df, rsa_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, rsa_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, rsa_df, on="Artist", how="left")
rnr4_df = pd.merge(rnr4_df, rsa_df, on="Artist", how="left")
```

```python
#merge hot 100 singles data to df's on artist
hot_df = pd.read_csv("data/hot100singles.csv")
hot_df = hot_df.rename(columns={"Row Labels" : "Artist"})
rnr_df = pd.merge(rnr_df, hot_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, hot_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, hot_df, on="Artist", how="left")
rnr4_df = pd.merge(rnr4_df, hot_df, on="Artist", how="left")
```

```python
#merge Rolling Stone covers data to df's on artist
rsc_df = pd.read_csv("data/RSCovers.csv")
rnr_df = pd.merge(rnr_df, rsc_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, rsc_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, rsc_df, on="Artist", how="left")
rnr4_df = pd.merge(rnr4_df, rsc_df, on="Artist", how="left")
```

```python
#merge All Music average album scores to df's on artist
abs_df = pd.read_csv("data/Album_scores.csv")
rnr_df = pd.merge(rnr_df, abs_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, abs_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, abs_df, on="Artist", how="left")
rnr4_df = pd.merge(rnr4_df, abs_df, on="Artist", how="left")
```

```python
#merge spotify artist play data to df's on artist
spt_df = pd.read_csv("data/spotify.csv")
rnr_df = pd.merge(rnr_df, spt_df, on="Artist", how="left")
rnr2_df = pd.merge(rnr2_df, spt_df, on="Artist", how="left")
rnr3_df = pd.merge(rnr3_df, spt_df, on="Artist", how="left")
rnr4_df = pd.merge(rnr4_df, spt_df, on="Artist", how="left")
```

```python
#replace Nan's with 0's
rnr_df.fillna(0, inplace=True)
rnr2_df.fillna(0, inplace=True)
rnr3_df.fillna(0, inplace=True)
rnr4_df.fillna(0, inplace=True)
```

```python
#Add In/Out of RnR Hall of Fame column to df's 
rnr_df["In_Out"] = 1
rnr2_df["In_Out"] = 0
rnr3_df["In_Out"] = 0
rnr4_df["In_Out"] = 0
```

```python
#combine inductees and snubs dataframes to use as train/test data
train_df = pd.concat ([rnr_df, rnr2_df]).reset_index(drop=True)
```

```python
#output df's to csv's
train_df.to_csv("output/artist_data.csv")
rnr3_df.to_csv("output/2019.csv")
rnr4_df.to_csv("output/2044.csv")
```
